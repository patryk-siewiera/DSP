## SINUS

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from scipy import signal
import scipy.fftpack
```

## TRIANGLE

```python
A = np.power(2,16)/2-1 # 16 bit, max
fs = 44100
f = 10
t = np.linspace(0, 1, 500)
sine_wave=A*np.sin(2*np.pi*t*f)


plt.plot(sine_wave) # print 1 second
```

```python
t = np.linspace(0, 1, 500)
triangle = A * signal.sawtooth(2 * np.pi * 5 * t, 0.5)
plt.plot(t, triangle)
```

## SQUARE

```python
t = np.linspace(0, 1, 500)
square = A * signal.square(2 * np.pi * 5 * t, 0.5)
plt.plot(t, square)
```
