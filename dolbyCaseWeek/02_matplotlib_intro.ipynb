```python
# generating values
xvals = np.arange(-20,20, 0.01) # cz,as / przedziedzial czasowy
yvals = np.cos(xvals)
```

```python
# formattting
plt.title('Title of plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axis([-20,20,-1.5,1.5]) #[xmin, xmax, ymin, ymax]
plt.plot(xvals,yvals)
```

```python
A = np.power(2,16)/2-1 # 16 bit, max
fs = 44100
f = 10
t = np.arange(-5,5,1./fs) # -time, time, step

sine_wave=A*np.sin(2*np.pi*t*f)

sine = sine_wave.astype(np.int16)
sec = fs

plt.plot(sine[: sec]) # print 1 second
```
