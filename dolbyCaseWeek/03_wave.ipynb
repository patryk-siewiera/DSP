```python
import wave
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

#player audio in jupyter
import IPython.display as ipd 
```

```python
# open a file
wf = wave.open('bongos.wav', mode= 'rb')
```

```python
wf.getnchannels() 
```

```python
wf.getsampwidth() 
```

```python
wf.getframerate()
```

```python
wf.getnframes()
```

```python
A = np.power(2,16)/2-1
fs = 44100
f = 1000
t = np.arange(-5,5,1./fs)

sine_wave=A*np.sin(2*np.pi*t*f)
sine = sine_wave.astype(np.int16)

# 2 channels
sine_2ch=np.repeat(sine,2)

plt.plot(sine[:1000])
```

```python
wav_out= wave.open(r'out.wav',mode='wb')

wav_out.setnchannels(2)

wav_out.setsampwidth(2)

wav_out.setframerate(fs)

wav_out.writeframes(sine_2ch)
```

```python
ipd.Audio('out.wav')
```
